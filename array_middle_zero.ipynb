```python
import numpy as np

# Create 1D array from 1 to 20
arr = np.arange(1, 21)

# Reshape to a 4x5 matrix
mat = arr.reshape(4, 5)

# Set the middle element to 0 (uses floor center for even dimensions)
center_row = mat.shape[0] // 2
center_col = mat.shape[1] // 2
mat[center_row, center_col] = 0

# Print the resulting matrix
print(mat)
```
