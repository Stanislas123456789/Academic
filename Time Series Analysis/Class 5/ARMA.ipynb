```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.arima_process import ArmaProcess
```

```python
# Define ARMA(2,2) process parameters
ar_params = np.array([1, 0.4, 0.1])  # AR coefficients (including the lag term)
ma_params = np.array([1, -0.5, 0.3])   # MA coefficients (including the lag term)
```

```python
# Generate an ARMA(2,2) process
arma_process = ArmaProcess(ar_params, ma_params)
simulated_data = arma_process.generate_sample(nsample=200)
```

```python
# Plot the simulated time series
time = np.arange(len(simulated_data))
plt.figure(figsize=(12, 5))
plt.plot(time, simulated_data, label='ARMA(2,2) Process')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Simulated ARMA(2,2) Time Series')
plt.legend()
plt.show()
```

```python
# Plot ACF and PACF
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sm.graphics.tsa.plot_acf(simulated_data, lags=30, ax=axes[0])
axes[0].set_title('Autocorrelation Function (ACF)')
sm.graphics.tsa.plot_pacf(simulated_data, lags=30, ax=axes[1])
axes[1].set_title('Partial Autocorrelation Function (PACF)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define AR and MA polynomials
ar_params = np.array([1, 0.4, 0.25])  # AR polynomial coefficients
ma_params = np.array([1, -0.5, 0.25]) # MA polynomial coefficients

# Compute the roots of AR and MA polynomials
ar_roots = np.roots(ar_params)
ma_roots = np.roots(ma_params)

# Plot the roots on the complex plane
plt.figure(figsize=(6,6))
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')

plt.scatter(ar_roots.real, ar_roots.imag, color='red', label='AR Roots', marker='o')
plt.scatter(ma_roots.real, ma_roots.imag, color='blue', label='MA Roots', marker='x')

# Draw the unit circle
circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='dashed', linewidth=1.2)
plt.gca().add_patch(circle)

plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.title('Roots of AR and MA Polynomials')
plt.legend()
plt.grid(True, linestyle='dotted')

# Show the plot
plt.show()

# Return the numerical values of the roots
ar_roots, ma_roots
```
