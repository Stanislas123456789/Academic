```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import periodogram
from statsmodels.graphics.tsaplots import plot_acf

# Gnration de la srie temporelle AR(1)
np.random.seed(42)
n = 200  # Nombre de points
alpha2 = 0.8  # Coefficient AR
sigma = 1.0  # cart-type du bruit
y = np.zeros(n)
y[0] = np.random.normal(0, sigma)  # Valeur initiale

for t in range(1, n):
    y[t] = alpha2 * y[t-1] + np.random.normal(0, sigma)

# Convertir en srie pandas
time_series = pd.Series(y)

# Tracer la srie temporelle AR
plt.figure(figsize=(10, 6))
plt.plot(time_series, label="Srie AR gnre")
plt.title("Srie Temporelle AR Gnre")
plt.xlabel("Temps")
plt.ylabel("Valeur")
plt.legend()
plt.show()

# Calculer et tracer l'ACF
plt.figure(figsize=(10, 6))
plot_acf(time_series, lags=20, alpha=None, title="Fonction d'Autocorrlation (ACF)")
plt.show()

# Calculer le priodogramme
fs = 1  # Frquence d'chantillonnage (1 observation par unit de temps)
frequencies, power_spectrum = periodogram(time_series, fs=fs)

# Tracer le priodogramme
plt.figure(figsize=(10, 6))
plt.plot(frequencies, power_spectrum, color='b')
plt.title("Priodogramme du Processus AR")
plt.xlabel("Frquence (cycles par unit de temps)")
plt.ylabel("Densit spectrale de puissance")
plt.grid(True)
plt.show()
```
