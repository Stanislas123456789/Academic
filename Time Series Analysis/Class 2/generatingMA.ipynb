```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Parameters
np.random.seed(42)  # Set seed for reproducibility
n = 500  # Length of time series
mu = 0  # Mean of the process
sigma = 1  # Standard deviation of white noise
q = 2  # Order of moving average process
theta = [0.9, 0.7]  # MA coefficients

# Generate white noise
white_noise = np.random.normal(mu, sigma, n)

# Generate moving average process
ma_process = np.zeros(n)
for t in range(q, n):
    ma_process[t] = mu + white_noise[t] + sum(theta[i] * white_noise[t - i - 1] for i in range(q))

# Create DataFrame
df = pd.DataFrame({
    'White Noise': white_noise,
    'Moving Average Process': ma_process
})

# Plot results
plt.figure(figsize=(12, 5))
plt.plot(df['White Noise'], label='White Noise', linestyle='dashed', alpha=0.7)
plt.legend()
plt.title('White Noise')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

```python
# Plot results
plt.figure(figsize=(12, 5))
plt.plot(df['Moving Average Process'], label='Moving Average Process', linewidth=2)
plt.legend()
plt.title('Moving Average Process')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.tsa.stattools as acf
from statsmodels.graphics.tsaplots import plot_acf # For ACF plot  
```

```python
plt.figure(figsize=(10, 5))
plot_acf(df['White Noise'], lags=50, alpha=0.1)
plt.title('White Noise ACF')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plot_acf(df['Moving Average Process'], lags=50, alpha=0.1)
plt.title('Moving Average Process ACF')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()
```
