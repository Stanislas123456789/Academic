```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg

# Generate synthetic time series data (AR(1) process)
np.random.seed(42)
n = 200  # Number of data points
alpha = 0.8  # AR coefficient
sigma = 1.0  # Standard deviation of noise
y = np.zeros(n)
y[0] = np.random.normal(0, sigma)  # Initial value

for t in range(1, n):
    y[t] = alpha * y[t-1] + np.random.normal(0, sigma)

# Convert the generated data into a pandas Series
time_series = pd.Series(y)

# Plot the time series
plt.figure(figsize=(10, 6))
plt.plot(time_series, label="Generated Time Series")
plt.title("Generated Time Series (AR(1) Process)")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf
from statsmodels.graphics.tsaplots import plot_acf
```

```python
# Plot ACF
plt.figure(figsize=(10,5))
plot_acf(time_series, lags=50)  # Adjust lags as needed
plt.title("Autocorrelation Function (ACF) of AR process")
plt.show()
```
